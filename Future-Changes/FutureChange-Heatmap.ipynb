```python
#import netCDF4 as nc
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
fig = plt.figure(figsize=(12,11))
ax = fig.subplots(nrows=4, ncols=3)
var = ('tmax','t95','days_above_t95','t05','tmin','days_below_t05','frostdays','precipitation',\
       'p95','days_above_p95','wetdays')
# Loop through each variable
for vv in np.arange(np.shape(var)[0]):
    print(var[vv])
#Read data as pandas dataframe
    df = pd.read_csv(f'./Future_changes_{var[vv]}.csv',nrows=6) 
    df = df.set_index('Experiment')

    if vv==0:
        ax[0,0].set_title(var[vv])
        ax[0,0].set_aspect("equal")
        bplot1   = sns.heatmap(df,ax=ax[0,0],cmap="hot_r",vmin=1,vmax=2.5,xticklabels=False);
    if vv ==1:
        ax[0,1].set_title(var[1])
        ax[0,1].set_aspect("equal")
        bplot1   = sns.heatmap(df,ax=ax[0,1],cmap="hot_r",vmin=0,vmax=5,xticklabels=False,yticklabels=False)
    if vv ==2:
        ax[1,0].set_title(var[2])
        ax[1,0].set_aspect("equal")
        bplot1   = sns.heatmap(df,ax=ax[1,0],cmap="hot_r",vmin=10,vmax=30,xticklabels=False)
    if vv ==3:
        ax[1,1].set_title(var[3])
        ax[1,1].set_aspect("equal")
        bplot1   = sns.heatmap(df,ax=ax[1,1],cmap="hot_r",vmin=0,vmax=5,xticklabels=False,yticklabels=False)
    if vv ==4:
        ax[1,2].set_title(var[4])
        ax[1,2].set_aspect("equal")
        bplot1   = sns.heatmap(df,ax=ax[1,2],cmap="hot_r",vmin=1.0,vmax=2.5,yticklabels=False,xticklabels=False)
    if vv ==5:
        ax[2,0].set_title(var[5])
        ax[2,0].set_aspect("equal")
        bplot1   = sns.heatmap(df,ax=ax[2,0],cmap="hot",vmin=-15,vmax=-5,xticklabels=False)
    if vv ==6:
        ax[2,1].set_title(var[6])
        ax[2,1].set_aspect("equal")
        bplot1   = sns.heatmap(df,ax=ax[2,1],cmap="hot",vmin=-30,vmax=-10,yticklabels=False,xticklabels=False)
    if vv==7:
        ax[2,2].set_title(var[7])
        ax[2,2].set_aspect("equal")
        bplot1   = sns.heatmap(df,ax=ax[2,2],cmap="YlGnBu",vmin=0,vmax=0.5,yticklabels=False,xticklabels=False);
    if vv==8:
        ax[3,0].set_title(var[8])
        ax[3,0].set_aspect("equal")
        bplot1   = sns.heatmap(df,ax=ax[3,0],cmap="YlGnBu",vmin=0,vmax=5)
    if vv==9:
        ax[3,1].set_title(var[9])
        ax[3,1].set_aspect("equal")
        bplot1   = sns.heatmap(df,ax=ax[3,1],cmap="YlGnBu",vmin=0,vmax=2,yticklabels=False)
    if vv==10:
        ax[3,2].set_title(var[10])
        ax[3,2].set_aspect("equal")
        bplot1   = sns.heatmap(df,ax=ax[3,2],cmap="YlGnBu",vmin=-2,vmax=8,yticklabels=False)
fig.delaxes(ax[0,2])
plt.savefig('heatmap_allchange1.pdf')
```
